```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import datetime
from egaznepy.visualize import apply_plot_style
apply_plot_style(1.5)
plt.rcParams['text.usetex'] = False
# Read CSV files for Xcorr model
df_xcorr_keele = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_xcorr_with_lpc_keele.csv', header=None, names=['RCA'])
df_xcorr_fda = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_xcorr_with_lpc_fda.csv', header=None, names=['RCA'])
df_xcorr_mocha_timit = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_xcorr_with_lpc_mocha_timit.csv', header=None, names=['RCA'])
df_xcorr_timit = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_xcorr_with_lpc_timit.csv', header=None, names=['RCA'])

# Read CSV files for IF model
df_if_keele = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_if_keele.csv', header=None, names=['RCA'])
df_if_fda = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_if_fda.csv', header=None, names=['RCA'])
df_if_mocha_timit = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_if_mocha_timit.csv', header=None, names=['RCA'])
df_if_timit = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_if_timit.csv', header=None, names=['RCA'])

#Read CSV files for Xcorr model without LPC
df_xcorr_keele_wo_lpc = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_xcorr_wo_lpc_keele.csv', header=None, names=['RCA'])
df_xcorr_fda_wo_lpc = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_xcorr_wo_lpc_fda.csv', header=None, names=['RCA'])
df_xcorr_mocha_timit_wo_lpc = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_xcorr_wo_lpc_mocha_timit.csv', header=None, names=['RCA'])
df_xcorr_timit_wo_lpc = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_xcorr_wo_lpc_timit.csv', header=None, names=['RCA'])


df_joint_keele = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_joint_keele.csv', header=None, names=['RCA'])
df_joint_fda = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_joint_fda.csv', header=None, names=['RCA'])
df_joint_mocha_timit = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_joint_mocha_timit.csv', header=None, names=['RCA']) # CHECK AGAIN
df_joint_timit = pd.read_csv('/dataHDD/chsaikeerthi/2024-chirumamilla/results/rca_per_utt_joint_timit.csv', header=None, names=['RCA'])

# Add model and dataset labels
df_xcorr_keele['Model'] = 'Xcorr'
df_xcorr_keele['Dataset'] = 'KEELE'
df_xcorr_fda['Model'] = 'Xcorr'
df_xcorr_fda['Dataset'] = 'FDA'
df_xcorr_mocha_timit['Model'] = 'Xcorr'
df_xcorr_mocha_timit['Dataset'] = 'MOCHA-TIMIT'
df_xcorr_timit['Model'] = 'Xcorr'
df_xcorr_timit['Dataset'] = 'TIMIT'

df_if_keele['Model'] = 'IF'
df_if_keele['Dataset'] = 'KEELE'
df_if_fda['Model'] = 'IF'
df_if_fda['Dataset'] = 'FDA'
df_if_mocha_timit['Model'] = 'IF'
df_if_mocha_timit['Dataset'] = 'MOCHA-TIMIT'
df_if_timit['Model'] = 'IF'
df_if_timit['Dataset'] = 'TIMIT'

df_xcorr_keele_wo_lpc['Model'] = 'Xcorr_wo_lpc'
df_xcorr_keele_wo_lpc['Dataset'] = 'KEELE'
df_xcorr_fda_wo_lpc['Model'] = 'Xcorr_wo_lpc'
df_xcorr_fda_wo_lpc['Dataset'] = 'FDA'
df_xcorr_mocha_timit_wo_lpc['Model'] = 'Xcorr_wo_lpc'
df_xcorr_mocha_timit_wo_lpc['Dataset'] = 'MOCHA-TIMIT'
df_xcorr_timit_wo_lpc['Model'] = 'Xcorr_wo_lpc'
df_xcorr_timit_wo_lpc['Dataset'] = 'TIMIT'

df_joint_keele['Model'] = 'Joint'
df_joint_keele['Dataset'] = 'KEELE'
df_joint_fda['Model'] = 'Joint'
df_joint_fda['Dataset'] = 'FDA'
df_joint_mocha_timit['Model'] = 'Joint'
df_joint_mocha_timit['Dataset'] = 'MOCHA-TIMIT'
df_joint_timit['Model'] = 'Joint'
df_joint_timit['Dataset'] = 'TIMIT'

# Combine all data
df = pd.concat([
    df_xcorr_keele, df_xcorr_fda, df_xcorr_mocha_timit, df_xcorr_timit,
    df_if_keele, df_if_fda, df_if_mocha_timit, df_if_timit,
    df_xcorr_keele_wo_lpc, df_xcorr_fda_wo_lpc, df_xcorr_mocha_timit_wo_lpc, df_xcorr_timit_wo_lpc,
    df_joint_keele, df_joint_fda, df_joint_mocha_timit, df_joint_timit
])

# Plotting
plt.figure(figsize=(12, 7))

# Create violin plots
sns.violinplot(data=df, x='Dataset', y='RCA', hue='Model', cut=0,  bw=0.3)

# Add title and labels
plt.title('RCA Comparison by Model and Dataset')
plt.xlabel('Dataset')
plt.ylabel('RCA')
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')

# Save the plot
now = datetime.datetime.now()
results_dir = '/dataHDD/chsaikeerthi/2024-chirumamilla/results1'
if not os.path.exists(results_dir):
    os.makedirs(results_dir)

#plt.savefig(f'{results_dir}/violin_plot_comparison_{now.strftime("%Y-%m-%d_%H-%M-%S")}.pdf', bbox_inches='tight')

filename = f'{results_dir}/violin_plot_{datetime.datetime.now().strftime("%Y%m%d_%H%M%S")}.pdf'
try:
    plt.savefig(filename, format='pdf', bbox_inches='tight')
    print(f'Plot saved successfully to {filename}')
except Exception as e:
    print(f'Error saving plot: {e}')
plt.show()
```
